```python
import tensorflow as tf

#함수직접 만들어보자.

#키와 신발의 상관관계

height = [170]      #여러개 넣어줌 키=[170, ...]
shoe = [260]        #여러개 넣어줌

#shoe = a*height + b 라고 선형으로 예측

a = tf.Variable(0.1)    #텐서 플로우로 변수 만드는 문법
b = tf.Variable(0.2)    #초기값을 0.2로 시작

def costFunction():     #최소제곱법으로 만든 손실함수
    Expectation = (height*a + b)        #예측모델
    return tf.square(260-Expectation)       #제곱함수
    #손실함수 만드는 메소드 중 하나. 손실값 = 오차(실제값 - 예측값)

opt = tf.keras.optimizers.Adam(learning_rate = 0.1)     #경사하강법으로 a(w)를 업데이트 해주는 문법(a(x)를 보통 w로 나타낸다.)
      #텐서플로우 안에 있는 케라스 라이브러리 안에서
      #optimizer(최적화)로 경사하강법 사용
      #adam() 기본값으로는 0.001 정도임. learning_rate = 0.1. 0.1씩 업데이트.
      #Adam이 기울기를 알아서 learning_rate 안써도 Adam이 알아서 스마트하게 업데이트 해줌

      #경사하강법 이용해서 가중치 자동업데이트해주는 고마운 도구
      #learning_rate 는 얼마만큼 한번에 변수들을 업데이트 할지 숫자 지정. learning_rate의 조절에 따라 값이 달라지므로 결과가 잘 나오도록 값을 잘 찾아주는것이 매우 중요
      #var_list[a,b] : 경사하강법으로 업데이트할 variable 목록

for i in range(100):        #여러번 학습
    opt.minimize(costFunction, var_list=[a,b])      #경사하강법으로 업데이트 할 변수 목록을 업데이트 해서 최적값 찾음
    #opt.minimize(비용(손실)함수, var_list[a,b]) var_list[a,b] : 경사하강법으로 업데이트할 variable 목록
    print( a.numpy(), b.numpy())        #print(a, b) 쓸데없는 정보까지 출력
    #텐서는 .numpy() 메서드(method)를 호출하여 NumPy 배열로 변환할 수 있습니다
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf

from google.colab import drive
drive.mount('/content/drive')

#판다스에서 가져오는거라 dataframe이다
data = pd.read_csv('/content/drive/MyDrive/gpascore.csv')
data.head()
```

admit 0 불합격, 1 합격, gre 점수, gpa 학점, rank 대학등급

```python
#데이터 전처리(머신러닝 돌리기전에 pre-processing)
print(data.isnull().sum())
```

```python
#na 인 애들 다 삭제하고
data = data.dropna()
#그 값들 출력
print(data.isnull().sum())
```

데이터 읽어오고, 0(na)인거 확인하고 결측치 다 날려버리고 필요한것만 가져온다. 그것이 데이터 전처리(머신러닝을 돌리기 직전 상태로 만들어 놓은거)

```python
data.describe
```

```python
y = data['admit'].values
x=[]

for i, rows in data.iterrows():
    #iterows() generates an iterator object of DataFrame, allowing us to iterate each row in DataFrame(from Pandas)
    #각 행의 정보
    print(rows['gre'], rows['gpa'], rows['rank'])
    x.append( [ rows['gre'], rows['gpa'], rows['rank'] ] )
```

```python
x
```

```python
#텐서플로우를 이용한 선형회귀(딥 러닝을 이용한 선형회귀)

#예측은 기계학습한 모델들이 하는거다.

#모델을 만들자! 딥러닝 모델로. 딥러닝 모델에는 은익층이 필요하다. 많아도 되는데 일단 3개만 만들자.
#은익층을 순서대로 만들거다. 딥러닝 모델을 만드는데 Sequential 을 사용하면 layers 로 층을 만든다. 맨끝의 메모리용량에는 1이 들어간다.
#인수의 갯수가 마지막엔 1개가 되야한다.
#tf.keras.layers.Dense(메모리용량 mb, activation = 함수(각각의 레이어층의 property 들을 어떠한 가중치를 가지고 판단할 것인가) )

#은익층을 어떻게 설계하겠다.
model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(64, activation='tanh'),
        tf.keras.layers.Dense(128, activation='tanh'),
        tf.keras.layers.Dense(1, activation = 'sigmoid'),       #마지막이어도 끝날때는 콤마(,)를 써줘야한다. 문법이니까
])

model.compile(optimizer= 'adam', loss = 'binary_crossentropy', metrics='accuracy')
#             최적화(경사하강법)에서는 adam을 많이 쓰고
#                                손실함수(cost function)에서 0,1 두개로 판단할때는 binary_crossentropy를 제일 자주 사용한다.
#                                                              메트릭스(metrics)를 '정확도'로 판단하겠다.
#                                                              메트릭스 안에 값을 넣을건데 '정확도'로 값을 넣겠다. 정확하게 맞췄는지 틀렸는지 찾겠다.



#accuracy : 예를들어서 1000명의 환자중 990명이 음성이다. 암을 판별하는 인공지능을 만들거다(구라로).
#무조건 다 음성만 뜨게 해서 기계를 만들면 맞을 확률이 990/1000 = 99%다. 이거 사기다.




#fit학습시키는 명령어. x,y를 리스트로 만들어서 array로 만들어서 넣어야한다. array가 속도 더 빠르게 해준다.
#epochs : 몇번을 실행할것인가. 학습실행횟수. 경사하강에서 몇번 돌릴지 정하는 학습 실행 횟수
model.fit(np.array(x), np.array(y), epochs=1000)
```

```python
#gre 700, 학점 3.7 rank 4대학원 합격확률예측

Prediction = model.predict([ [750, 3.70, 3], [400,2.2,3]])
print(Prediction)
```
