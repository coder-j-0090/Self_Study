#라벨인코딩

라벨 인코딩(Label Encoding)은 범주형 데이터를 수치형으로 변환하는 기법 중 하나입니다. 머신 러닝 알고리즘은 일반적으로 숫자로 된 데이터를 처리하므로, 범주형 데이터를 숫자로 인코딩하여 학습 모델에 입력할 수 있습니다.

라벨 인코딩은 주로 명목형 데이터(카테고리 간 순서가 없는 데이터)를 변환하는 데 사용됩니다. 예를 들어, "red", "blue", "green"과 같은 색상을 갖는 데이터가 있다면, 이를 라벨 인코딩을 통해 숫자로 변환할 수 있습니다.

간단한 라벨 인코딩의 예시:

"red" -> 0
"blue" -> 1
"green" -> 2
라벨 인코딩은 범주의 수에 따라 숫자를 할당하며, 라벨 간의 순서를 반영하지 않습니다. 따라서 머신 러닝 모델이 숫자의 크기로 판단하는 알고리즘(예: 선형 회귀)에는 적합하지 않을 수 있습니다.

주의할 점은 라벨 인코딩을 적용할 때, 범주형 데이터의 값들에 대해 고유한 정수 값을 할당해야 합니다. 또한, 라벨 인코딩은 명목형 데이터에만 사용되어야 하며, 순서를 가지는 순서형 데이터(예: 나이 그룹)에는 적합하지 않습니다.

대안으로 원-핫 인코딩(One-Hot Encoding)을 사용하여 범주형 데이터를 처리할 수 있습니다. 원-핫 인코딩은 각 범주를 이진 특성으로 변환하며, 범주에 해당하는 특성에만 1을 할당하고 나머지는 0을 할당합니다. 이렇게 변환된 데이터는 각 범주 간에 독립적인 특성으로 인식되어 선형 모델 등에 적용하기에 적합합니다.

#라벨링

머신 러닝에서 라벨링(Labeling)은 지도 학습(Supervised Learning)의 한 단계로, 데이터셋의 각 샘플에 대해 정답으로 사용될 레이블(라벨)을 부여하는 과정을 말합니다.

지도 학습은 주어진 입력 데이터와 해당 데이터에 대한 출력(정답) 데이터(라벨) 사이의 관계를 학습하는 머신 러닝 방법입니다. 이때 학습에 사용되는 데이터셋에는 입력 데이터와 그에 해당하는 출력 데이터가 함께 포함되어 있어야 합니다. 이러한 출력 데이터가 레이블로, 머신 러닝 모델이 학습하는 과정에서 입력 데이터와 레이블 간의 관계를 학습하게 됩니다.

예를 들어, 스팸 메일 필터링 모델을 구축한다고 가정해보겠습니다. 스팸 메일 필터링 모델을 학습시키기 위해서는 스팸과 일반 메일을 구분하는 레이블(1 또는 0)을 메일 데이터에 부여해야 합니다. 이때, 스팸 메일에 대해서는 레이블 1을, 일반 메일에 대해서는 레이블 0을 부여합니다. 그리고 학습 과정에서 모델은 메일의 특징과 레이블 간의 관계를 학습하여 새로운 메일이 스팸인지 아닌지를 예측할 수 있게 됩니다.

라벨링은 지도 학습에서 매우 중요한 단계이며, 정확한 레이블링은 모델의 성능에 큰 영향을 미칩니다. 라벨링된 데이터셋을 사용하여 모델을 학습시키고 검증하면서, 모델이 레이블에 따라 정확한 예측을 수행하도록 하는 것이 머신 러닝의 목표입니다.

```python
#라벨링 : 종, 섬, 성별이 문자로 되어있어서 머신러닝이 쉽도록 숫자로 변환(라벨링)해 주는것
#종속변수(y=결과값)을 라벨링이라고 부른다.

from sklearn.preprocessing import LabelEncoder      #라벨 인코딩을 위한 라이브러리
label = ['species', 'island', 'sex']
#컴퓨터가 각 항목의 갯수를 카운트해서 자동으로 숫자로 라벨링(변환)시켜준다. 라벨링 한거를 다시 저장한다.
df[label] = df[label].apply(LabelEncoder().fit_transform)
df.head()
```

#더미처리

머신 러닝에서 더미 처리(Dummy Encoding)는 범주형 변수를 수치형으로 변환하는 방법 중 하나입니다. 범주형 변수는 카테고리 형태의 값을 가지는 변수로, 머신 러닝 모델에 바로 적용할 수 없으므로 수치형으로 변환해야 합니다. 더미 처리는 이러한 범주형 변수를 0과 1로 이루어진 이진 변수로 변환하는 방법입니다.

더미 처리 과정은 다음과 같습니다:

범주형 변수를 해당 카테고리에 속하는지 여부를 나타내는 이진 변수로 변환합니다.
원래 변수의 카테고리 수 만큼 새로운 이진 변수를 생성합니다.
각 샘플은 해당 카테고리에 속하는 경우 1로 표시되고, 아닌 경우 0으로 표시됩니다.
예를 들어, "색상"이라는 범주형 변수에 "빨강", "파랑", "녹색" 세 가지 카테고리가 있다고 가정하겠습니다. 더미 처리를 하면 "색상" 변수는 "빨강", "파랑", "녹색" 세 개의 이진 변수로 변환됩니다. 각 샘플은 자신이 속한 색상 카테고리에 대응하는 이진 변수가 1이고, 나머지는 0인 형태가 됩니다.

더미 처리를 통해 범주형 변수를 수치형으로 변환하면, 머신 러닝 모델에 적용하기 쉽습니다. 또한, 더미 처리를 통해 각 카테고리 간의 순서가 없는 범주형 변수를 선형 모델과 같은 숫자를 요구하는 알고리즘에 적용할 수 있습니다.

```python
#데이터 변환처리, 더미처리
#데이터 변환처리 : 섬하고 성별(두개다 중간값이 없다.)은 이산변수(두개다 중간값이 없다.) 이므로(연속형 변수 아님).

#이산변수를 카테고리로 변수로 바뀐다.

#더미처리 : 컴퓨터에서는 0하고 1로 처리한다.
#특성값들을 여러개 깔아놓고 각 특성들이 영향을 미쳣냐 아닌가를 비교해서 학습시키는게 빠르다. 이것이 더미처리다.
#유치원생이 초콜릿, 사탕, 홍삼캔디 묶음을 구매하려면 홍삼캔디가 싫은데 다른 2개는 좋아서 상쇄도 되고 복잡해지는데
#초콜릿, 사탕, 홍상캔디가 따로따로 있으면 어떤게 좋고 싫은지 구분하기 좋다.

#마키노차야

#요약
#데이터 타입을 카테고리화해서 더미처리를 해줘야 각 특성값 별로 분류해준다.
import pandas as pd
category =['island', 'sex']     #이산 변수이므로 범주에 해당하는 카테고리형으로 변환, species 는 종속변수이므로 일단 생략
for i in category:
    df[i] = df[i].astype('category')        #각 섹션별로 카테고리화 시켜서 분류시킨다.
df = pd.get_dummies(df) #더미처리는 각각의 범주에 대해 1,0 으로 처리.
```

```python
df.head()
```

```python
#파생변수 만들기
#연속범위의 값들을 구간화 시켜서 표현

#자동으로 범위구간을 나우어서 자르라. 최댓값 최솟값 찾아서 N(여기서는 5)개의 구간으로 나눠라

#요약
#연속범위의 값들은 구간화 시켜서 파생변수를 만들어주는게 낫다. 연속범위는 더미처리 안해줘도 된다(몸무게는 모든 펭귄이 다 존재한다. 그래서 그대로 두는게 맞다)
df['body_mass_g_qcut'] = pd.qcut(df['body_mass_g'], 5, labels = False) #구간값 5
df.head()
```

```python
df['body_mass_g_qcut'].value_counts().sort_index()
```

```python
#스케일 작업 : 여러 데이터들이 단위가 다를경우 맞추어 주어야 함
from sklearn.preprocessing import MinMaxScaler      #0-1사이의 숫자로 바꿔줌
scaler = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
min = MinMaxScaler()
min.fit(df[scaler])     #scale 데이터에 맞게 만들고
df[scaler] = min.transform(df[scaler])  #scaler에 맞게 데이터 변환
```

```python
df.head()
```

```python
df.tail()
```

```python
#데이터 분리

from sklearn.model_selection import train_test_split
#학습을 시켜서 테스트할 때 사용한다.
X_train, X_test, y_train, y_test = train_test_split( df.iloc[:, 1:], df['species'], test_size = 0.2, stratify = df['species'], random_state = 1)
#                                                                                                   stratify 종류에 따라 특정종이 몰리지 않게 골고루 분포.
#                                                                                                   random_state=1 을 사용해서 데이터를 랜덤으로 섞어준다.
```
