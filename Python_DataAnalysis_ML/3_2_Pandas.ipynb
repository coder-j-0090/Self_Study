# 데이터프레임

```python
#size = (x길이, y길이)
rawData = np.random.randint(50,100,size = (4,3))
rawData
```

```python
#미리 만들어 져 있는 거에다가 인덱스를 줄거다
#표로 만들기
df = pd.DataFrame(rawData,
                  index = ['1반','2반','1반','2반'],
                  columns=['국','영','수'])
df
```

```python
df['국']
```

```python
df[0]   #인덱스가 이미 지정되 있는데 0으로 지정되지 않아서 에러가 생긴다.
```

```python
#딕셔너리에서 없는 키값을 입력하면 만들어진다.
#round는 반올림 함수

df['평균'] = round( (df['국'] + df['수'] + df['영'])/3, 2)
df
```

```python
df['na'] = np.nan    #결측치 추가해보기. 결측치도 숫자로 인정
df
```

```python
del df['na'] #삭제
df
```

```python
#boolean indexing
df[ df.평균 > 75]
```

```python
#필요없는 애들은 지우고 쓸 수 있다.
df = df.drop(['평균'], axis = 'columns')
df
```

```python
#결측치 하기 위해서 타입을 float 라고 넣었다

df = df.astype('float64')
df
```

```python
df.dropna(axis = 0)     #결측치 있는 줄(가로)을 지워라
```

```python
df
```

```python
df.dropna(axis=1)   #결측치 있는 줄(세로)을 지워라
```

```python
df.fillna('hello')
```

```python
df.fillna(0)
```

```python
#평균으로 값을 채워놔야지 평균값이나 그 외 값에 미치는 영향이 적다
#fillna(n) = fill not a number with n
df.fillna(df.mean())
```

# 값의 연결

```python
a = pd.DataFrame(np.arange(1,10).reshape(3,3))
a
```

```python
b = pd.Series(np.arange(10,40,10))
b
```

a&b 를 연결시켜주자

```python
#concatenate : 연결
pd.concat([a,b],axis=1)
```

```python
#원래 인덱스가 있는데 원래 0이었다. 근데 보기 싫으니까 원래있던 인덱스 무시하고 차례로 다시붙여라.
pd.concat([a,b], axis = 1, ignore_index = True)
```

```python
a.append(b, ignore_index = True)
```
