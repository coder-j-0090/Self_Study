```python
#sklearn(싸이킷런)에서 iris 데이터 셋을 제공해준다.
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#딕셔너리 형태로 저장된 데이터들의 키값을 출력해준다
iris.keys()
```

```python
iris.data[:5]
```

```python
iris.feature_names
```

```python
X = iris.data[:, 2:]
```

```python
#target 은 y(종속, 결과)
iris.target[:5]
```

```python
import numpy as np
```

```python
#unique가 중복되지 않고 유일한거의 종류가 3개다(0,1,2). 그리고 그거의 갯수가 각각 50개씩 들어있다. knn에서는 shape 으로 array의 갯수를 판별한다.
#여기서는 unique 사용

#결측치 고를때는 data.isnull().sum()

np.unique(iris.target, return_counts = True)
```

```python
iris.target_names                   #dtype은 여기선느 문자열 10개라는 거다
```

```python
y = iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier     #의사결정나무(분류기)로 위의 3중 어느것인지 판별하겠다(setosa, versicolor, virginica)
tree = DecisionTreeClassifier(max_depth = 2)        #깊이를 2층으로 만들겠다. max_depth가 중요한 하이퍼 파라미터다. 깊이가 너무 깊으면 속도 느려지고 값 이상해짐
#학습시키기
tree.fit(X,y)
```

없어도 된다. 시각화 파트니까. 머신러닝이랑은 별 관계 없다.

```python
#트리형태로 표현해주는 기능
from sklearn.tree import export_graphviz
```

```python
export_graphviz(tree, out_file = "iris.dot", feature_names = iris.feature_names[2:],
                class_names = iris.target_names, rounded=True, filled=True, impurity = True)
```

```python
import pydot
#파이썬에서 dot 스크립트 언어를 사용하기위한 도구
#graphviz는 내부적으로 dot 라는 스크립트 언어 사용
```

```python
graph = pydot.graph_from_dot_file("iris.dot")[0]
```

```python
iris_png = graph.create_png()
```

```python
from IPython.core.display import Image
Image(iris_png)

#자동으로 트레이닝하면서 바꿔준다.
```
